```python
import squarify
import matplotlib.pyplot as plt

# font size 설정
plt.rcParams["font.size"] = 16

solved_prob=[40,120,20,60,60]
name=['yujin','hamjin','swan','wuyong','changyun']
 
squarify.plot(sizes=solved_prob, label=name, value = solved_prob,alpha=.8)
plt.axis('off')
plt.show()
```
